## 5.Checking outliers.
An outlier  is a data point that significantly deviates from the majority of the dataset. It is an observation that lies outside
the expected range or distribution, often because of variability in the data, measurement errors, or anomalies.

Characteristics of Outliers:
They are far from the mean or median of the data.
They can skew statistical analyses, leading to misleading results.
Identifying and handling outliers is critical in data preprocessing to ensure accurate model performance.

```python

import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns

# Box plot for the height of employees
plt.figure(figsize=(8, 6))
sns.boxplot(
    df_employee['Height'], 
    vert=True, 
    patch_artist=True, 
    boxprops=dict(facecolor="lightgreen", color="darkgreen"),
    whiskerprops=dict(color="blue", linestyle="--"),
    capprops=dict(color="purple"),
    medianprops=dict(color="red", linewidth=2),
    flierprops=dict(marker="o", color="orange", markersize=8)
)
plt.title("Box Plot of Employee Heights at ABC Company", fontsize=14, color="darkblue")
plt.ylabel("Height (cm)", fontsize=12, color="darkred")
plt.show()
```

## Investigate if there's any correlation between age and salary, and represent it visually. 

CORRELATION:

Values close to +1 indicates strong positive correlation, -1 indicates a strong negative correlation and 0 indicates suggests no linear correlation. Darker colors signify strong correlation, while light colors represents weaker correlations. Positive correlation variable move in same directions. As one increases, the other also increases. Negative correlation variable move in opposite directions. An increase in one variable is associated with a decrease in the other.HEAT MAPSMaps

Heat Maps is a type of plot which is necessary when we need to find the dependent variables. One of the best way to find the relationship between the features can be done using hea.mapsIn

```python
df3 = df_employee[["Age","Salary"]]
df3
```

```python
#Compute correlation
corr = df3.corr()
corr
```

```python
#Create Heat Map
sns.heatmap(corr,annot=True,cmap="coolwarm")
#Add tittle
plt.title("Correlation Matrix Between Age And Salary Of Employees")
#Show plot
plt.show()
```

DATA STORY :
The heatmap provides actionable insights by quantifying the strength and direction of relationships, with annotations highlighting exact correlation values and a color gradient visually reinforcing these trends.

Key Findings from the Heatmap
Age vs. Salary:

A strong positive correlation (e.g., 0.7 or above) between Age and Salary indicates that employees with more years of experience or seniority earn higher salaries. This is often typical in organizations with hierarchical pay scales where tenure plays a significant role in compensation.

A weak or negligible correlation (e.g., close to 0) would suggest that salary is not significantly influenced by age. This could reflect modern compensation strategies focused on meritocracy or skill-based pay rather than tenure.
